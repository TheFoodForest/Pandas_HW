```python
import pandas as pd
import numpy as np

schools_df = pd.read_csv('schools_complete.csv')
students_df = pd.read_csv('students_complete.csv')

display(schools_df.head())
display(students_df.head())
```

District Summary

```python
total_schools = len(schools_df['school_name'])
total_students = len(students_df['student_name'])
total_budget = schools_df['budget'].sum()
average_math = students_df['math_score'].mean()
av_read = students_df['reading_score'].mean()
district_sum = pd.DataFrame({
    'Total Schools':[total_schools],
    'Total Students':[total_students],
    'Total Budget':[total_budget],
    'Average Math Score':[average_math],
    'Average Reading Score':[av_read]
})
num_pas_math = len(students_df.loc[students_df['math_score']>60,'math_score'])
num_pas_read = len(students_df.loc[students_df['reading_score']>60,'reading_score'])
district_sum['% Passing Math'] = round((num_pas_math/total_students)*100,2)
district_sum['% Passing Reading'] = round((num_pas_read/total_students)*100,2)
district_sum['% Passing Rate'] = round((district_sum['% Passing Math']+district_sum['% Passing Reading'])/2,2)
district_sum
```

School Summary

```python
#this for loop takes quite some time to complete
for i in range(len(students_df)):
    if students_df.loc[i,'reading_score'] > 60:
        students_df.loc[i,'pass_reading'] = 1
    if students_df.loc[i,'math_score'] > 60:
        students_df.loc[i,'pass_math'] = 1

#display(students_df.loc[students_df['math_score']<60,:].head())



group_school = students_df.groupby('school_name')
school_type = schools_df.loc[:,['school_name','type','budget']]
total_students = pd.DataFrame(group_school['student_name'].count())
avg_math = pd.DataFrame(group_school['math_score'].mean())

avg_read = pd.DataFrame(group_school['reading_score'].mean())
num_math = pd.DataFrame((group_school['pass_math']).count())
num_reading = pd.DataFrame(group_school['pass_reading'].count())
school_sum = school_type.merge(total_students.merge(avg_math.merge(avg_read.merge(num_math.merge(num_reading, on='school_name'),on='school_name'),on='school_name'),on='school_name'),on='school_name')
```

```python
school_sum['% Passing Math'] = round((school_sum['pass_math']/school_sum['student_name'])*100,2)
school_sum['% Passing Reading'] = 100 
school_sum['Overall Passing Rate(%)'] = round((school_sum['% Passing Math'] + school_sum['% Passing Reading'])/2,2)
school_sum.drop(columns=['pass_math','pass_reading'], inplace=True)
school_sum.rename(columns={
    'type':'school_type',
    'student_name':'count_of_students',
    'math_score':'avg_math_score',
    'reading_score':'avg_reading_score'
},inplace=True)
```

```python
print('School Summary')
display(school_sum)
```

```python
school_sum = school_sum.sort_values('Overall Passing Rate(%)', ascending=False)
top5 = school_sum.iloc[0:5,:]
top5
```

Summary By Budget Per Student

```python
full_df = students_df.merge(schools_df, on='school_name')
```

```python
full_df['Budget Per Student'] = full_df['budget']/full_df['size']
maxb = full_df['Budget Per Student'].max()
minb = full_df['Budget Per Student'].min()
print(maxb)
print(minb)
print((maxb-minb)/4)
```

```python
bins = [0,580,600,620,640,660]
labels = ['<580','580-599','600-619','620-639','>640']
full_df['budget_group'] = pd.cut(full_df['Budget Per Student'], bins, labels=labels)
full_df.head()
```

```python
budget_grouped = full_df.groupby('budget_group')
avg_math = pd.DataFrame(budget_grouped['math_score'].mean())
avg_read = pd.DataFrame(budget_grouped['reading_score'].mean())
num_math = pd.DataFrame(budget_grouped['pass_math'].count())
count_stu = pd.DataFrame(budget_grouped['Student ID'].count())
budget_sum = avg_math.merge(avg_read.merge(num_math.merge(count_stu, on='budget_group'),on='budget_group'),on='budget_group')
budget_sum['% Passing Math'] = round((budget_sum['pass_math']/budget_sum['Student ID'])*100,2)
budget_sum['% Passing Reading'] = 100
budget_sum['Passing Rate (%)'] = round((budget_sum['% Passing Math']+100)/2,2)
budget_sum.drop(columns=['Student ID', 'pass_math'], inplace=True)
print('Summary By Budget Per Student')
budget_sum
```

Summary By School Size 

```python
max_size = full_df['size'].max()
min_size = full_df['size'].min()
print(max_size)
print(min_size)
print((max_size-min_size)/3)
```

```python
bins = [0,1500,3000,5000]
labels = ['Small','Medium','Large']
full_df['size_group'] = pd.cut(full_df['size'], bins, labels=labels)
full_df.head()
```

```python
size = full_df.groupby('size_group')

avg_math = pd.DataFrame(size['math_score'].mean())
avg_read = pd.DataFrame(size['reading_score'].mean())
num_math = pd.DataFrame(size['pass_math'].count())
count_stu = pd.DataFrame(size['Student ID'].count())
size_sum = avg_math.merge(avg_read.merge(num_math.merge(count_stu, on='size_group'),on='size_group'),on='size_group')
size_sum['% Passing Math'] = round((size_sum['pass_math']/size_sum['Student ID'])*100,2)
size_sum['% Passing Reading'] = 100
size_sum['Passing Rate (%)'] = round((size_sum['% Passing Math']+100)/2,2)
size_sum.drop(columns=['Student ID', 'pass_math'], inplace=True)
print('Summary By School Size')
size_sum
```
